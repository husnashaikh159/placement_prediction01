```python
import pandas as pd
```

```python
df=pd.read_csv('Student_dataset.csv')
```

```python
df.info()
```

```python
# Null Value Checking
df.isna().sum()
```

```python
# seperating input and output feature
x = df.drop('placement_status', axis=1)
y = df[['placement_status']]
```

```python
# preprocessing
num=[]
cat=[]
for i in x.columns:
  if x[i].dtypes!=object:
    num.append(i)
  else:
    cat.append(i)
xcat=x[cat]
xnum=x[num]
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
le=LabelEncoder()
scaler = StandardScaler()
```

```python
for i in xcat.columns:
  xcat[i]=le.fit_transform(xcat[i])
```

```python
xnum = pd.DataFrame(scaler.fit_transform(xnum),columns=num)
```

```python
X=xnum.join(xcat)
```

```python
import pickle
with open('scaler.pkl','wb') as file1:  # wb is for write binary
    pickle.dump(scaler,file1)
```

```python
xnum
```

```python
# dividing training and testing data
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytest.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfr=RandomForestClassifier(n_estimators=25)
```

```python
model=rfr.fit(xtrain,ytrain)
```

```python
tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
tr_score=accuracy_score(ytrain,tr_pred)
ts_score=accuracy_score(ytest,ts_pred)
```

```python
print('Training score:',tr_score)
print('Testing score:',ts_score)
```

```python
# confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tr_mat=confusion_matrix(ytrain,tr_pred)
ts_mat=confusion_matrix(ytest,ts_pred)
```

```python
print('Training:')
print(tr_mat)
print('Testing:')
print(ts_mat)
```
